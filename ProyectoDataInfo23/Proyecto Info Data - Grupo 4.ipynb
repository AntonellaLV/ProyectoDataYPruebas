```python
# Importar bibliotecas
import requests
import json
from datetime import datetime
import pandas as pd
```

```python
# Obtener una clave de API de OpenWeatherMap # Construir la URL de la API, # Realizar la solicitud HTTP a la API
# Guardar la respuesta en un archivo JSON # Imprimir los datos del clima

def construir_url(api_key, latitud, longitud):
    # Construir la URL de la API con la latitud, longitud y clave de API
    url = f'https://api.openweathermap.org/data/2.5/weather?lat={latitud}&lon={longitud}&appid={api_key}&lang=sp&units=metric'
    return url

def realizar_solicitud(url):
    # Realizar la solicitud HTTP a la API
    response = requests.get(url)
    if response.status_code == 200:
        # Si la respuesta es exitosa, obtener el contenido JSON de la respuesta
        data = response.json()
        # Guardar los datos en un archivo JSON con el nombre 'datos_clima.json'
        with open("datos_clima.json", 'w') as outfile:
            json.dump(data, outfile)
        return data
    else:
        print(f"Error al hacer la solicitud: Código de estado {response.status_code}")
        return None

def imprimir_datos_clima(response_json):
    # Imprimir los datos del clima contenidos en response_json
    print("Datos del clima:")
    print(f"Ciudad: {response_json['name']}")
    print(f"Temperatura: {response_json['main']['temp']}°C")
    print(f"Sensación térmica: {response_json['main']['feels_like']}°C")
    print(f"Humedad: {response_json['main']['humidity']}%")
    print(f"Descripción del clima: {response_json['weather'][0]['description']}")
    print(f"Velocidad del viento: {response_json['wind']['speed']} m/s")

if __name__ == '__main__':
    # Agregar la clave de API de OpenWeatherMap
    api_key = 'cf9cc3cd9216d61286cb5b7ffa89d392'  # Reemplaza con tu clave de API de OpenWeatherMap

    # Coordenadas geográficas de la ubicación deseada (puedes cambiarlos por otros valores)
    latitud = -27.451100
    longitud = -58.986600

    # Construir la URL de la API
    url = construir_url(api_key, latitud, longitud)

    # Realizar la solicitud y obtener los datos del clima
    response_json = realizar_solicitud(url)

if response_json is not None:
    # Imprimir los datos del clima
    imprimir_datos_clima(response_json)

```

```python
if response_json is not None:
        # Normalizar los datos JSON contenidos en la variable 'response_json'
        data_normalize = pd.json_normalize(response_json)
        weather = pd.json_normalize(response_json['weather'])

        # Crear DataFrames a partir de los datos normalizados
        s1 = pd.DataFrame(data_normalize)
        s2 = pd.DataFrame(weather)

        # Concatenar los DataFrames en uno solo a lo largo del eje horizontal (axis=1)
        df = pd.concat([s1, s2], axis=1)

        # Eliminar la columna 'weather'
        df.drop('weather', inplace=True, axis=1)

        # Agregar la fecha y hora actual al DataFrame
        df['fecha_hora'] = datetime.now()

        # Imprimir el DataFrame resultante
        print(df.to_string())

        # Guardar los datos en un archivo CSV
        df.to_csv('data_clima.csv', index=False)
```

**API Pokemon**

```python
# Cualquier Pokemon a elección

import requests
import pandas as pd
import json

def obtener_datos_pokemon(pokemon_name):
    url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}/'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        # Guardar los datos en un archivo JSON con el nombre 'pokemon_<nombre>.json'
        with open(f'pokemon_{pokemon_name.lower()}.json', 'w') as outfile:
            json.dump(data, outfile)
        return data
    else:
        print(f"Error al hacer la solicitud: Código de estado {response.status_code}")
        return None

if __name__ == '__main__':
    # Obtener el nombre del Pokémon del usuario
    pokemon_name = input("Ingresa el nombre o número del Pokémon: ")

    # Obtener información sobre el Pokémon especificado por el usuario
    data = obtener_datos_pokemon(pokemon_name)

    if data is not None:
        # Crear un DataFrame a partir de los datos del Pokémon
        df = pd.DataFrame({
            'Nombre': [data['name']],
            'Peso': [data['weight']],
            'Altura': [data['height']],
            'Experiencia Base': [data['base_experience']],
            'Habilidades': [', '.join([habilidad['ability']['name'] for habilidad in data['abilities']])],
            'Tipos': [', '.join([tipo['type']['name'] for tipo in data['types']])],
            'Movimientos': [', '.join([movimiento['move']['name'] for movimiento in data['moves']])]
        })

        # Imprimir los datos del Pokémon en formato pandas
        print(df.to_string(index=False))

        # Guardar los datos en un archivo CSV
        df.to_csv(f'pokemon_{pokemon_name.lower()}.csv', index=False)
```

**"API Star Wars"**

```python
import requests
import pandas as pd
import json

def obtener_datos_personaje(personaje_id):
    url = f'https://swapi.dev/api/people/{personaje_id}/'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        # Guardar los datos en un archivo JSON con el nombre 'personaje_<personaje_id>.json'
        with open(f'personaje_{personaje_id}.json', 'w') as outfile:
            json.dump(data, outfile)
        return data
    else:
        print(f"Error al hacer la solicitud: Código de estado {response.status_code}")
        return None

if __name__ == '__main__':
    # Obtener información sobre el personaje con ID 1 (Luke Skywalker)
    personaje_id = 1
    data = obtener_datos_personaje(personaje_id)

    if data is not None:
        # Crear un DataFrame a partir de los datos del personaje
        df = pd.DataFrame({
            'Nombre': [data['name']],
            'Altura': [data['height']],
            'Peso': [data['mass']],
            'Color de pelo': [data['hair_color']],
            'Color de piel': [data['skin_color']],
            'Color de ojos': [data['eye_color']],
            'Año de nacimiento': [data['birth_year']],
            'Género': [data['gender']],
            'Planeta de origen': [data['homeworld']],
            'Películas': [', '.join(data['films'])],
            'Naves espaciales': [', '.join(data['starships'])],
            'Vehículos': [', '.join(data['vehicles'])],
            'URL': [data['url']]
        })

        # Imprimir los datos del personaje en formato pandas
        print(df.to_string(index=False))

        # Guardar los datos en un archivo CSV
        df.to_csv(f'personaje_{personaje_id}.csv', index=False)
```
